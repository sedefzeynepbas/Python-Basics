```python
plt.subplot(2,2,1)
plt.plot(x,y,"blue")
plt.subplot(2,2,2)
plt.plot(y,x,"yellow")
plt.subplot(2,2,3)
plt.plot(y,x,"red")
plt.subplot(2,2,4)
plt.plot(x,x**2,"black")
plt.show()
```

```python
fig = plt.figure(figsize=(6,4))
axes= fig.add_axes([0,0,1,1])
axes.plot(x,x** 0.5,"red",label = "X Karekök")
axes.plot(x,x**2,"pink",label= "X Kare")
axes.plot(x,x**3,"#8080ff",label = "X Küp")#hex color codu :****
axes.legend()#küçük renk kodunu bu kod yapıyor
```

```python
#grafikleri özelleştirme başlıyor...
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**2,"red",linewidth=3,ls="--",marker="o",markersize=10,markerfacecolor="black",markeredgecolor="yellow",markeredgewidth=3)
axes.set_xlim(0,10)
axes.set_ylim(-1,40)
#linewidth : line kalınlığı değişiyor, lw = de yazabilirsin
#linestyle = ls ->"":"" yaparsan da noktalı olur ya "-."
#marker : değerleri işaretliyor "o" mesela küçük o ile
#markersize: noktaların rengini değiştirir
#markerfacecolor: renk değiştirir
#edge marker kenar rengi
#size marker kenar boyu
```
