```python
import numpy as np
import pandas as pd
```

```python
#concatenation->ekleme->iki dataseti
```

```python
dataset1 = {
    "A": ["A1","A2","A3","A4"],
    "B":["B1","B2","B3","B4"],
    "C":["C1","C2","C3","C4"],
}
```

```python
dataset2 = {
    "A": ["A5","A6","A7","A8"],
    "B":["B5","B6","B7","B8"],
    "C":["C5","C6","C7","C8"],
}
```

```python
df1 = pd.DataFrame(dataset1,index = [1,2,3,4])
df2 = pd.DataFrame(dataset2,index = [5,6,7,8] )
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2])#columnlar aynı topladık...
```

```python
pd.concat([df1,df2],axis=1)
```

```python
pd.concat([df1,df2],axis=1).fillna(value=0)
```

```python
#join sql left join-> benziyor...
```

```python
dataset1 = {
    "A" : ["A1","A2","A3","A4"],
    "B" : ["B1","B2","B3","A4"],
}
dataset2 = {
    "X" : ["X1","X2","X3"],
    "Y" : ["Y1","Y2","Y3"],
    
}
df1 = pd.DataFrame(dataset1,index = [1,2,3,4]) 
df2 = pd.DataFrame(dataset2,index = [1,2,3])
```

```python
df1#join işleminde buradaki tüm indeksler gelir
```

```python
df2
```

```python
df1.join(df2)#outer, inner ve left join var. Buradaki left join
```

```python
df2.join(df1)
```

```python
dataset1 = {
    "A" : ["A1","A2","A3"],
    "B" : ["B1","B2","B3"],
    "anahtar" : ["K1","K2","K3"]
}
```

```python
df1 = pd.DataFrame(dataset1,index = [1,2,3]) 
```

```python
dataset2 = {
    "X" : ["X1","X2","X3","X4"],
    "Y" : ["Y1","Y2","Y3","Y4"],
    "anahtar" : ["K1","K2","K5","K4"]
}
```

```python
df2 = pd.DataFrame(dataset2,index = [1,2,3,4])
```

```python
df1
```

```python
df2
```

```python
#inner join gibi, sadece ortak kısımları birleştirir(anahtara göre yaparsan sadece k1 ve k2nin olduğu satırlar gelecektir)
```

```python
pd.merge(df1,df2)
```
